#### example 1: initialize

- kaiming init(HE init)
    + kaiming_normal() and kaiming_uniform_()
    + fan_in: number of neurons in input layer
    + *applied with ReLU
- xavier init(Glorot init)
    + xavier_normal() and xavier_uniform_()
    + fan_in & fan_out

```python
import torch
from torch import nn

linear1 = nn.Linear(5, 3)
nn.init.constant_(linear1.weight, 3)
print(linear1.weight.data)

linear2 = nn.Linear(5, 3)
nn.init.kaiming_uniform_(linear2.weight, 3)
print(linear2.weight.data)

linear3 = nn.Linear(5, 3)
nn.init.xavier_uniform_(linear3.weight, 3)
print(linear3.weight.data)

```

#### example 2: build neuron network

```python
import torch
from torch import nn
from torchsummary import summary

class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden1 = nn.Linear(3, 3)
        self.hidden2 = nn.Linear(3, 2)
        self.output = nn.Linear(2, 2)

        nn.init.xavier_normal_(self.hidden1.weight)
        nn.init.zeros_(self.hidden1.bias)
        nn.init.kaiming_normal_(self.hidden2.weight)
        nn.init.zeros_(self.hidden2.bias)

    def forward(self, x):
        x = torch.sigmoid(self.hidden1(x))
        x = torch.relu(self.hidden2(x))
        x = torch.softmax(self.output(x), dim=-1)
        return x
    

if __name__ == '__main__':
    model = MyModel()
    data = torch.randn(5, 3)
    print(f'data = {data}')
    output = model.forward(data)
    print(f'output = {output}')
    print('parameter:')
    summary(model, input_size=(5, 3))
```

#### example 3: loss function

- loss function = cost func = objective func = error func
- designed for evaluating fitting condition
- classification problem:
    + CrossEntropyLoss
    + BCELoss (Binary Category Entropy)
- regression problem:
    + MAE: mean absolute error (not smooth)
    + MSE: mean squared error (gradient explosion might occur)
    + Smooth L1: combine MAE and MSE
