## chapter2: tensor & numpy

#### example1

```python
import torch
import numpy

t1 = torch.randint(1, 10, (2, 3))
n1 = t1.numpy()
print(f't1 = {t1}, n1 = {n1}\n')
print(f'type of n1 = {type(n1)}\n')

t2 = torch.randint(1, 10, (2, 3))
n2 = t2.numpy().copy()
t2[0] = 255
print(f't2 = {t1}, n2 = {n2}\n')
print(f'type of n2 = {type(n2)}\n')

n3 = numpy.array([[1,2,3], [4,5,6]])
t3 = torch.from_numpy(n3)
print(f't3 = {t3}, n3 = {n3}\n')

num1 = torch.tensor(1).item()
print(f'num1 = {num1}, type = {type(num1)}\n')
```

#### example2

```python
import torch

t1 = torch.tensor([1, 2, 3, 4, 5])
tmp1 = torch.tensor([[1,2,3],
                     [4,5,6]])
tmp2 = torch.tensor([[1,2],
                     [3,4],
                     [5,6]])

t2 = torch.add(t1, t1)
print(t2)
t3 = t1 + 10
print(t3)
t4 = t1.neg()
print(t4)
t5 = torch.matmul(t1, t1)
print(t5)
t6 = tmp1 @ tmp2
print(t6)

```

#### example3

```python
import torch

tmp1 = torch.tensor([
    [1, 2, 3],
    [4, 5, 6]
    ])

print(tmp1.exp())
print(tmp1.log())
print(f'min = {tmp1.min()}, max = {tmp1.max()}')
print(f'sum0 = {tmp1.sum(0)}, sum1 = {tmp1.sum(1)}, sum = {tmp1.sum()},')
print(tmp1.float().mean())
```

#### example5

```python
import torch

w = torch.tensor(10, requires_grad=True, dtype=torch.float)
loss = 2 * w ** 2
loss.backward()
w.data = w.data - 0.01 * w.grad
print(f'w_new = {w}')
```

```python
import torch

x = torch.tensor(data=10, requires_grad=True, dtype=torch.float)
for i in range (1, 1000):
    loss = x ** 2 + 20
    loss.backward()
    x.data -= 0.01 * x.grad
    x.grad.zero_()
print(f'x = {x}, y = {loss}')
```
