### chapter1: tensor

#### example1

```python
import torch
import numpy as np

t1 = torch.tensor(10)
print(f't1 = {t1}, type = {type(t1)}\n')

t2 = torch.tensor([1,2,3,4,5])
print(f't2 = {t2}, type = {type(t2)}\n') 

t3 = torch.tensor(np.random.randint(0, 10, size = (2, 3)))
print(f't3 = {t3}, type = {type(t3)}\n') 

t4 = torch.Tensor(10)
print(f't4 = {t4}, type = {type(t4)}\n') 
#大写可以直接创建
```

#### example2

```python
import torch

t1 = torch.ones(2, 3)
print(f't1 = {t1}, type = {type(t1)}\n')

t2 = torch.zeros(2, 3)
print(f't2 = {t2}, type = {type(t2)}\n')

tmpTensor = torch.tensor([1,2,3])
t3 = torch.zeros_like(tmpTensor)
print(f't3 = {t3}, type = {type(t3)}\n')

t4 = torch.full((2,3), 255)
print(f't4 = {t4}, type = {type(t4)}\n')

t5 = torch.full_like(tmpTensor, 255)
print(f't5 = {t5}, type = {type(t5)}\n')
```

#### example3

```python
import torch

tmp = torch.tensor((2, 3))

t1 = torch.randint(0, 10, (2, 3))
print(f't1 = {t1}\n')

t2 = torch.randint_like(tmp, 0, 10)
print(f't2 = {t2}\n')

t3 = torch.arange(0, 20, 3)
print(f't3 = {t3}\n')

t4 = torch.linspace(0, 20, 19)
print(f't4 = {t4}\n')
```

#### example4

```python
import torch

t1 = torch.tensor([1.1 ,2,3,4,5], dtype=torch.float16)
print(f't1 = {t1}, dtype = {t1.dtype}\n')
t2 = t1.type(torch.int16)
print(f't1 = {t2}, dtype = {t2.dtype}\n')
```
